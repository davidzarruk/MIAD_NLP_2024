# FastText

Importar embeddings preentrenados y reducir dimensionalidad:

```python
import fasttext
import fasttext.util
ft = fasttext.load_model('cc.es.300.bin')
fasttext.util.reduce_model(ft, 20)
ft.get_dimension()
ft.save_model('cc.es.20.bin')
```

Embeddings de algunas palabras:

```python
import matplotlib.pyplot as plt
from gensim.models import FastText

# Load the FastText model
model = FastText.load_fasttext_format('cc.es.20.bin')

# Define words we want to compare
words = ['reina', 'mujer', 'nia', 'nio', 'hombre', 'rey', 'reina', 'agua']

# Plot the embeddings of the words
fig, axs = plt.subplots(1, len(words), figsize=(len(words)*1, 10))  # Double the width

for i, word in enumerate(words):
    axs[i].imshow(model.wv[word].reshape(20, 1), aspect='auto')  # 'auto' is the default
    axs[i].set_title(word)

plt.show()
```

```python
# Perform word analogy calculation
word = 'rey'
result = model.wv.most_similar(word, topn=10)

# Print the result
print(f"Similar words to {word}:")
for word, similarity in result:
    print(f"{word}: {similarity}")
```

```python
model.wv.most_similar(positive=['rey', 'mujer'], negative=['hombre'], topn=10)
```
