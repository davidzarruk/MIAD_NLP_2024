Por medio de este notebook deber tokenizar el texto del set de noticias populares de UCL. ### Importar base de datos y libreras

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# Importacin libreras
import pandas as pd
import numpy as np
import scipy as sp
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
from nltk.stem.snowball import SnowballStemmer
%matplotlib inline
```

```python
# Carga de datos de archivos .csv
df = pd.read_csv('https://raw.githubusercontent.com/albahnsen/MIAD_ML_and_NLP/main/datasets/mashable_texts.csv', index_col=0)
df.head()
```

### Crear varaible de inters

```python
# Separacin de variable de inters (y)
y = df.shares
y.describe()
```

```python
# Categoizacin de la variable de inters (y)
y = pd.cut(y, [0, 893, 1200, 2275, 63200], labels=[0, 1, 2, 3])
y.value_counts()
```

```python
# Definicin de variable de inters en el dataframe
df['y'] = y
```

### Crear variables predictoras X_A - tokenizacin sin limpieza

```python
# Separacin de variables predictoras (X), solo se considera el texto de la noticia
X = df.text
```

```python
# Creacin de matrices de documentos usando CountVectorizer a partir de X
vect_A = CountVectorizer()
X_dtm_A = vect_A.fit_transform(X)
temp_A=X_dtm_A.todense()
```

```python
# Visualizacin de diccionario de palabras con su respectivo ID asignado
vect_A.vocabulary_
```

```python
# Impresin de dimensiones de matriz de documentos donde las filas son documentos y las columnas son trminos o tokens
X_dtm_A.shape
```

```python
# Visualizacin de 50 trminos en el diccionario de palabras
print(vect_A.get_feature_names()[-150:-100])
```

### Crear variables predictoras X_B - tokenizacin con limpieza de maysculas

```python
# Creacin de matrices de documentos usando CountVectorizer a partir de X, volviendo todas la palabras en minscula
# a partir del parmetro 'lowercase=False' 
vect_B = CountVectorizer(lowercase=False)
X_dtm_B = vect_B.fit_transform(X)
```

```python
# Impresin dimensiones de matriz de documentos donde las filas son documentos y las columnas son trminos o tokens
X_dtm_B.shape
```

```python
# Visualizacin de 50 trminos en el diccionario de palabras
print(vect_B.get_feature_names()[-150:-100])
```

### Crear variables predictoras X_C - tokenizacin con limpieza de maysculas y usando n-gramas

```python
# Creacin de matrices de documentos usando CountVectorizer a partir de X y usando n-gramas
# a partir del parmetro 'ngram_range=(1, 4)' 
vect_C = CountVectorizer(lowercase=False, ngram_range=(1, 4))
X_dtm_C = vect_C.fit_transform(X)
```

```python
# Impresin de dimensiones de matriz de documentos, donde las filas son documentos y las columnas son trminos o tokens
X_dtm_C.shape
```

```python
# Visualizacin de 50 trminos en el diccionario de palabras
print(vect_C.get_feature_names()[-150:-100])
```

###  Entrenar modelo de prediccin con diferentes matrices de palabras (variables predictoras)

```python
# Definicin de modelo Naive Bayes para predecir la varaible 'y' y variables predictoras x_A
nb = MultinomialNB()
pd.Series(cross_val_score(nb, X_dtm_A, y, cv=10)).describe()
```

```python
# Definicin de modelo Naive Bayes para predecir la varaible 'y' y variables predictoras x_B
nb = MultinomialNB()
pd.Series(cross_val_score(nb, X_dtm_B, y, cv=10)).describe()
```

```python
# Definicin de modelo Naive Bayes para predecir la varaible 'y' y variables predictoras x_B
nb = MultinomialNB()
pd.Series(cross_val_score(nb, X_dtm_C, y, cv=10)).describe()
```
